# Handling missing values

```python
numeric_cols = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
df.dropna(subset=numeric_cols, how='all', inplace=True)
```

```python
# Filling Missing Values using KNN technique
from sklearn.impute import KNNImputer
numeric_salaries = df.select_dtypes(include=[np.number])
knn_imputer = KNNImputer(n_neighbors=2)
imputed_data = knn_imputer.fit_transform(numeric_salaries)
imputed_df = pd.DataFrame(imputed_data, columns=numeric_salaries.columns)
```

```python
df[imputed_df.columns] = imputed_df
df.isnull().sum()
```

```python
df.info()
```

# Data processing

```python
# Convert dollar columns to numeric
currency_cols = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']
for col in currency_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Convert Year to datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y').dt.year
```

# Questions and EDA

### How would you detect and handle extreme outliers in TotalPayBenefits?

### How would you detect and handle extreme outliers in BasePay?

### Which statistical tests can you use to determine if TotalPayBenefits follows a normal distribution?

```python
#Salary Distributions & Outliers:

plt.figure(figsize=(12,6))
sns.histplot(df['TotalPayBenefits'], bins=50, kde=True)
plt.title('Distribution of Total Compensation')
plt.xlabel('Total Pay + Benefits ($)')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["TotalPayBenefits"], color='orange')

plt.xlabel("TotalPay + Benefits")
plt.title("Box Plot")
plt.show()
```

```python
Q1 = df["TotalPayBenefits"].quantile(0.25)
Q3 = df["TotalPayBenefits"].quantile(0.75)
IQR = Q3 - Q1  # Interquartile Range

# Outlier boundaries
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Filtering outliers
outliers = df[(df["TotalPayBenefits"] < lower_bound) | (df["TotalPayBenefits"] > upper_bound)]
print(f"Number of Outliers: {outliers.shape[0]}")
```

```python
print(outliers.sort_values(by="TotalPayBenefits", ascending=False).head(10))
```

```python
df["TotalPayBenefits"] = np.clip(df["TotalPayBenefits"], lower_bound, upper_bound)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["BasePay"], color='orange')

plt.xlabel("BasePay")
plt.title("Box Plot")
plt.show()
```

```python
Q1 = df["BasePay"].quantile(0.25)
Q3 = df["BasePay"].quantile(0.75)
IQR = Q3 - Q1  # Interquartile Range

# Outlier boundaries
lower_bounds = Q1 - (1.5 * IQR)
upper_bounds = Q3 + (1.5 * IQR)

# Filtering outliers
outliers = df[(df["BasePay"] < lower_bound) | (df["BasePay"] > upper_bound)]
print(f"Number of Outliers: {outliers.shape[0]}")
```

```python
df["BasePay"] = np.clip(df["BasePay"], lower_bounds, upper_bounds)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["BasePay"], color='orange')

plt.xlabel("BasePay")
plt.title("Box Plot")
plt.show()
```

### Can you find the top 5 job titles that have the highest variance in TotalPay?

### How does OvertimePay impact TotalPay across different JobTitle categories?

```python
# Group by JobTitle and calculate variance of TotalPay
job_variance = df.groupby("JobTitle")["TotalPayBenefits"].var()

# Sort in descending order and get top 5
top_variance_jobs = job_variance.sort_values(ascending=False).head(5)

# Display results
print("Top 5 Job Titles with Highest Variance in TotalPayBenefits:")
print(top_variance_jobs)
```

```python
top_variance_jobs.plot(kind="bar", figsize=(10,5), color="skyblue", edgecolor="black")
plt.ylabel("Variance in TotalPay")
plt.title("Top 5 Job Titles with Highest Salary Variance")
plt.xticks(rotation=45)
plt.show()
```

# MACHINE LEARNING(LINEAR REGRESSION)

```python
df['JobCategory'] = df['JobTitle'].str.split().str[0]

# Select features
features = ['Year', 'JobCategory', 'BasePay']
target = 'TotalPayBenefits'

# Preprocessing
# pd.get_dummies() converts categorical variables into numeric features.
# drop_first=True removes one dummy column to avoid redundancy and multicollinearity.
#Onehotencoding
df = df[features + [target]].dropna()
df = pd.get_dummies(df, columns=['JobCategory'], drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import warnings
X = df.drop(target, axis=1)
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
y_pred = model.predict(X_test)
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.2f}')
print(f'R² Score: {r2_score(y_test, y_pred):.2f}')
```
