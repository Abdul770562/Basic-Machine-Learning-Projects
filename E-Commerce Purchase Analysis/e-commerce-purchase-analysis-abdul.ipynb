# Q1) How many unique companies are present in the dataset? Which company has the highest number of transactions?

```python
# How many unique companies are present in the dataset? Which company has the highest number of transactions?
unique_company = df["Company"].nunique()
print(unique_company)

top_company = df["Company"].value_counts().idxmax()
print(top_company)

top_company_count = df["Company"].value_counts().max()
top_company_count
```

```python
df['Company'].value_counts().head(10).plot(kind='bar', title="Top 10 Companies by Transactions", color = "red")
plt.xlabel("Company Name")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()
```

# Q2) What is the distribution of purchase prices? Are there any outliers?


```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Purchase Price"], bins=30, kde=True, color='red')

plt.xlabel("Purchase Price")
plt.ylabel("Frequency")
plt.title("Distribution of Purchase Prices")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df["Purchase Price"], color='orange')

plt.xlabel("Purchase Price")
plt.title("Box Plot of Purchase Prices")
plt.show()
```

```python
Q1 = df["Purchase Price"].quantile(0.25)
Q3 = df["Purchase Price"].quantile(0.75)
IQR = Q3 - Q1  # Interquartile Range

# Outlier boundaries
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# Filtering outliers
outliers = df[(df["Purchase Price"] < lower_bound) | (df["Purchase Price"] > upper_bound)]
print(f"Number of Outliers: {outliers.shape[0]}")
```

 # Q3) Are there more transactions in the AM or PM? How does the time of purchase affect the purchase price?

```python
df.head()
```

```python
am_pm = df["AM or PM"].value_counts()
am_pm
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x=am_pm.index, y=am_pm.values, palette="coolwarm")

plt.xlabel("Time of Day")
plt.ylabel("Number of Transactions")
plt.title("AM vs. PM Transactions")
plt.show()
```

```python
am_pm_price = df.groupby("AM or PM")["Purchase Price"].mean()
print(am_pm_price)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="AM or PM", y="Purchase Price", data=df, palette="coolwarm")

plt.xlabel("Time of Day")
plt.ylabel("Purchase Price")
plt.title("Purchase Price Distribution in AM vs. PM")
plt.show()
```

#  Q4) How many customers made multiple purchases? What is the average number of purchases per unique customer?

```python
df.head(5)
```

```python
df["Email"].value_counts()
```

```python
customer_purchases = df["Email"].value_counts()
multiple_purchase_customers = customer_purchases[customer_purchases > 1].count()
multiple_purchase_customers
```

```python
# Total number of unique customers
total_unique_customers = customer_purchases.count()
print(total_unique_customers)

# Average purchases per unique customer
average_purchases = customer_purchases.mean()
print(average_purchases)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(customer_purchases, bins=30, kde=True, color='blue')

plt.xlabel("Number of Purchases")
plt.ylabel("Number of Customers")
plt.title("Distribution of Purchases per Customer")
plt.show()
```
